### Plotting the Data

```python
import matplotlib.pyplot as plt
```

```python
df_comp.spx.plot(figsize=(15,5), title = 'S&P500 Prices')
plt.show()
```

### The QQ Plot

```python
import scipy.stats
import pylab
```

```python
# Muestra como los datos se ajustan de manera normal
scipy.stats.probplot(df_comp.spx, plot= pylab)
pylab.show()
# Theorical quantiles representan cuantas desviaciones estandar se lejan de la media
# Linea roja representa los puntos que deben seguir los datos si estan distribuidos normalmente
```
